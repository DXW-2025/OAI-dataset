```python
import pandas as pd
from pathlib import Path

# ---------- 1. 路径 ----------
x_file  = r"C:\Users\DXW\Desktop\半月板部分切除术\OAI test\test1\X_cleaned.xlsx"
y_file  = r"C:\Users\DXW\Desktop\半月板部分切除术\OAI test\test1\Y_cleaned.xlsx"
id_col  = 'ID'
out_dir = Path(x_file).parent          # 结果放同目录

# ---------- 2. 读取 ----------
X = pd.read_excel(x_file, index_col=id_col)
Y = pd.read_excel(y_file, index_col=id_col)

# ---------- 3. 计算 ----------
# 将 X 和 Y 的列名先拼成 MultiIndex，再一次性算相关系数
corr_df = pd.DataFrame(
    {x_col: {y_col: X[x_col].corr(Y[y_col]) for y_col in Y.columns}
     for x_col in X.columns}
)

# 如果想让 X 列名在行、Y 列名在列，直接转置即可：
# corr_df = corr_df.T   # 按需开启

# ---------- 4. 保存 ----------
corr_df.to_csv(out_dir / 'X_vs_Y_pearson.csv')
corr_df.to_excel(out_dir / 'X_vs_Y_pearson.xlsx')

print('皮尔逊相关系数矩阵已保存至：')
print(out_dir / 'X_vs_Y_pearson.xlsx')
```
