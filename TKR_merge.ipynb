```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
add_prior_tkr.py  仅插入“prior to TKR”列并右移
用法:
    python add_prior_tkr.py 输入文件.xlsx 工作表名
"""
import sys
import pandas as pd

def main(file_path: str, sheet_name: str):
    # 读表
    df = pd.read_excel(file_path, sheet_name=sheet_name)

    # 在 G 列（索引 6）插入空列
    df.insert(6, 'prior to TKR', None)

    # 映射列名
    side_map = {
        1: 'RIGHT_CLOSEST OAI CONTACT PRIOR TO FOLLOW-UP KNEE REPLACEMENT',
        2: 'LEFT_CLOSEST OAI CONTACT PRIOR TO FOLLOW-UP KNEE REPLACEMENT'
    }

    # 填充新列
    df['prior to TKR'] = df.apply(
        lambda r: r[side_map.get(r['SIDE'], '')] if pd.notna(r.get('SIDE')) else None, axis=1
    )

    # 覆盖写回
    with pd.ExcelWriter(file_path, engine='openpyxl', mode='a',
                        if_sheet_exists='replace') as writer:
        df.to_excel(writer, sheet_name=sheet_name, index=False)

    print(f'已插入“prior to TKR”并右移，覆盖工作表“{sheet_name}”')

if __name__ == '__main__':
    main(r"C:\Users\DXW\Desktop\半月板部分切除术\OAI test\test 3\outcome.xlsx", "Sheet1")
```
